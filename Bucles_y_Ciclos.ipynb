## Bucle for El bucle for se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, conjunto o cadena de caracteres). Se ejecuta una vez por cada elemento en la secuencia.

# Suma
a = 10
b = 5
print(a + b)  # 15

# Resta
print(a - b)  # 5

# Multiplicacin
print(a * b)  # 50

# Divisin
print(a / b)  # 2.0

# Divisin entera
print(a // b)  # 2

# Mdulo
print(a % b)  # 0

# Exponenciacin
print(a ** b)  # 100000

# Asignacin compuesta
a += b
print(a)  # 15

# Redondeo
c = 3.14159
print(round(c, 2))  # 3.14

# Valor absoluto
d = -10
print(abs(d))  # 10

# Mximo y mnimo
e = [1, 2, 3, 4, 5]
print(max(e))  # 5
print(min(e))  # 1

# Potencia
print(pow(2, 3))  # 8

```python
# Imprimir todos los numeros de una secuencia de datos


secuencia= [1,2,3,4,5,]
# se lee " mientra la variable i este en la secunecia, haga esto... y el los imprimira a modo lista

for i in secuencia:
    print(i)
```

```python
# Clasificar los numeros que son pares y los que no son pares de la lista secuencia

for i in secuencia:
    if i % 2 == 0:
      print(f"El numero {i} es par ")
    else:
      print(f"El numero {i} es impar ")
```

## Supongamos que tienes una lista de precios de productos y quieres aplicar un descuento del 10% a cada precio.
## precios = [100, 150, 200, 250, 300]


```python
precio = [100,150,200,250,300]
precio_con_descuento = []


for i in precio:
    nuevo_precio = i*0.9
    precio_con_descuento.append(nuevo_precio)
    print(f"Precio full es: {i}, pero con descuento es: {nuevo_precio}")


```

```python
# Construir un contador que vaya hasta 5 y cuando la variable tenga este valor, deje de contar

contador = 0
while contador <5:
      print(contador)
      contador+=1
```

## Supongamos que quieres solicitar a un usuario que ingrese una contrasea hasta que lo haga correctamente.


```python
'''
break: Sale del bucle antes de que se evale la condicin o se termine de iterar sobre la secuencia.
continue: Salta el resto del cdigo dentro del bucle y vuelve a la condicin o a la siguiente iteracin.
else: Se ejecuta una vez cuando el bucle termina, a menos que se haya interrumpido con break.
'''

```

```python
contrasea_correcta="Segura123"
intentos = 3

while intentos >0:
      contrasea = input("Introduce tu contrasea: ")
      if contrasea == contrasea_correcta:
          print("Acceso concedido")
          break
      else :
          intentos-=1
          print(f"Contrasea incorrecta, te quedan {intentos} intentos")
if intentos == 0:
    print("Ya no te quedan intentos")

```
