## REGRESION LINEAL


```python
# Regresion lineal

# Grafica de dispersion: Sueldo vs Años de experiencia

# primero defino el grafico que qiuero hacer y uso la libreria

# eje y es Sueldo, eje x años de experiencia

x=salarios["Años de experiencia"]
y=salarios["Sueldo"]

# scatter es un tipo de graficp de dispersion, el primer valor es el eje x
# luego le eje y para generar la grafica

plt.scatter(x,y,color = "red")

# Titulo del grafico
plt.title("Sueldo vs Años de experiencia")

# Titulo eje x
plt.xlabel("Años de experiencia")
# Titulo eje y

plt.ylabel("Sueldo")

plt.show()

```
